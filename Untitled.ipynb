```python
%matplotlib inline
import time

import zarr
from distributed import Client
import dask.array as da
import numpy as np

clent = Client()

z = zarr.open("s3://aind-open-data/exaSPIM_730223_2025-04-25_14-44-58_flatfield-correction_2025-05-07_17-51-56/SPIM.ome.zarr/tile_000005_ch_561.zarr", 'r')
a = z['0']

squeeze = False

d = da.from_array(a, chunks=(1,1,128,256,256))

t0 = time.time()
if squeeze:
    d = d.squeeze()
    s = d[d.shape[0] // 2, ...].compute()
else:
    s = d[:, :, d.shape[2] // 2, ...].compute()
t1 = time.time()
print(f"squeeze {squeeze} ", t1-t0)
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.imshow(s.squeeze(), cmap='gray', aspect='auto', interpolation='nearest', vmin=0, vmax=50)
```
